## Task 2: Prediction using Unsupervised ML
From the given ‘Iris’ dataset, predict the optimum number of clusters
and represent it visually.

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
df = pd.read_csv('Iris.csv')
df.head() # first 5 rows
```

```python
x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
# The Elbow method for finding the optimum number of clusters
from sklearn.cluster import KMeans
pts=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    pts.append(kmeans.inertia_)
    
# plotting the graph of WCSS Vs Number of clusters

plt.figure(figsize=(10,5))
plt.title('WCSS Vs Number of clusters',color='black')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.plot(range(1,11),pts,color="black")

plt.show()
```

 In elbow method, the optimum number of clusters are where the elbow occurs.
 It happens when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.
 Elbow occurs at Number of Clusters=3
 Therefore we choose the Number of Clusters as 3

```python
# 
kmeans=KMeans(n_clusters=3)
y_pred=kmeans.fit_predict(x)
```

```python
# Visualizing Clusters on first 2 columns
df1=pd.DataFrame(x,y_pred)
plt.figure(figsize=(15,9))
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=75,c='red',label='Iris-setosa')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=75,c='blue',label='Iris-versicolor')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=75,c='green',label='Iris-Virginica')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',s=100,c='black',label='centroids')
plt.legend()
plt.show()
```
